```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
transit_gdf = gpd.read_file('Transit_Stops_for_King_County_Metro___transitstop_point.geojson')
```

```python
nbhds_gdf = gpd.read_file('Neighborhoods_Community_Reporting_Areas.geojson')
```

```python
transit_gdf.head()
```

```python
transit_gdf.plot()
```

```python
nbhds_gdf.plot()
```

```python
f, ax = plt.subplots()
nbhds_gdf.plot(ax=ax, facecolor='none', edgecolor='k')
transit_gdf.plot(ax=ax,  markersize=.2);
```

```python
#clip
transit_clip_gdf = transit_gdf.clip(nbhds_gdf)
```

```python
f, ax = plt.subplots()
nbhds_gdf.plot(ax=ax, facecolor='none', edgecolor='k')
transit_clip_gdf.plot(ax=ax,  markersize=.2);
```

```python
transit_clip_gdf.to_file('Seattle_Transit_Stops',driver='GeoJSON')
```
